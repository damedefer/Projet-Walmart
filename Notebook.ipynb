```python
Database['promotion_type'] = Database['promotion_type'].fillna('Ras')
```

```python
# Creer une colone ville et une autre departement
df = pd.DataFrame(Database)
# Séparer la colonne "location" en deux colonnes : ville et département
df[['Town', 'departement']] = df['store_location'].str.split(',', expand=True)
df = df.drop('store_location', axis = 1)
df
```

```python
#Utiliser GROUPBY afin de calculer le produit le prodiuit le plus vendu
Database_qtite_total = df.groupby("product_name")["quantity_sold"].sum().reset_index()
Database_qtite_total
```

```python
# Trouver le produit le plus vendu
produit_plus_vendu = Database_qtite_total.loc[Database_qtite_total["quantity_sold"].idxmax()]
print("Le produit le plus vendu est :", produit_plus_vendu["product_name"], 
      "avec", produit_plus_vendu["quantity_sold"], "ventes.")
```

```python
#Utiliser GROUPBY afin de calculer le produit le prodiuit le plus vendu en fonction des villes 
Database_qtite_total_town = df.groupby(["product_name", "Town"])["quantity_sold"].sum().reset_index()
Database_qtite_total_town
```

```python
produit_plus_vendu_par_ville = Database_qtite_total_town.loc[
    Database_qtite_total_town.groupby("Town")["quantity_sold"].idxmax()
]

print(produit_plus_vendu_par_ville)
```
